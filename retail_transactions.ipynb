# EXPLORATORY DATA ANALYSIS

### Distribution of Spend Across Payment Methods

```python
# Category-Level Analysis
# Distribution of spend across payment methods
paymentMethodTotalCost = df.groupby("Payment Method")["Total Cost"].agg(["sum", "mean", "min", "max"])
paymentMethodTotalCost.columns = ["Total Spend", "Avg. Spend", "Min. Spend", "Max. Spend"]
paymentMethodTotalCost = paymentMethodTotalCost.astype(int)

transactionTypes = {
    "Total Spend": paymentMethodTotalCost.iloc[0, 0:].values,
    "Avg. Spend": paymentMethodTotalCost.iloc[1, 0:].values,
    "Min. Spend": paymentMethodTotalCost.iloc[2, 0:].values,
    "Max. Spend" : paymentMethodTotalCost.iloc[3, 0:].values,
}


paymentMethods = ["Cash", "Credit Card", "Debit Card", "Mobile Payment"]
x = np.arange(len(paymentMethods))
width = 0.25
multiplier = 0


fig, axes = plt.subplots(figsize=(12, 8), dpi=100)

for transaction, values in transactionTypes.items():
  offset = width * multiplier
  rects = axes.bar(x + offset, values, width, label=transaction,
                   edgecolor="black")
  axes.bar_label(rects, padding=3)
  multiplier += 1

axes.set_xticks(x + width, paymentMethods)
axes.set_yscale("log")
axes.legend()
```

### Most and Less bought Products


```python
mostBoughtProducts = df.Product.value_counts(dropna=True)

productsUnique = []

for productList in df.Product:
  for product in productList.split(","):
    if product not in productsUnique:
      x = product.replace("[", "")
      x = x.replace("]", "")
      x = x.replace("'", "")
      x = x.strip()
      productsUnique.append(str(x))

productsUnique = list(productsUnique)
productsUnique = pd.Series(productsUnique)


# most and less bought products
mostBoughtProducts = productsUnique.value_counts(dropna=True).nlargest(10)
lessBoughtProducts = productsUnique.value_counts(dropna=True).nsmallest(10)
lessBoughtProducts.rename(index={"Feminine Hygiene Products": "Fem. Hygiene Prod."}, inplace=True)

fig, axes = plt.subplots(nrows=2, ncols=1, constrained_layout=True, figsize=(16, 10))

mostBar = axes[0].bar(mostBoughtProducts.index, mostBoughtProducts.values, ec="black", color="#F0CE41")
axes[0].bar_label(mostBar, label_type="center")

lessBar = axes[1].bar(lessBoughtProducts.index, lessBoughtProducts.values, ec="black", color="#68C2E8")
axes[1].bar_label(lessBar, label_type="center")

plt.suptitle("Most And Less Bought Products")
```

### Calculate average order value and CLV (Customer Lifetime Value)


```python
pd.set_option("display.max_columns", 15)
pd.set_option("display.max_rows", 15)
df["Average Order Value"] = df["Total Cost"] / df["Total Items"]

custLifetimeValue = df.groupby("Customer Name")["Average Order Value"].agg(["mean"])
custLifetimeValue.rename(columns={"mean": "Avg. CLV"}, inplace=True)
custLifetimeValue["Avg. CLV"] = custLifetimeValue.apply(lambda x: round(x, 2))

# plotting
largestCLV = custLifetimeValue["Avg. CLV"].nlargest(5)
lowestCLV = custLifetimeValue["Avg. CLV"].nsmallest(5)

xValuesFirst = np.arange(1, len(largestCLV.index) + 1, 1)
xValuesSecond = np.arange(1, len(custLifetimeValue.index) + 1, 1)

fig, ax = plt.subplots(figsize=(10, 8), dpi=100, nrows=2, ncols=1,
                       constrained_layout=True)

ax[0].plot(xValuesFirst, largestCLV, color="#6587E6", label="Largest", linestyle="--")
ax[0].plot(xValuesFirst, lowestCLV, color="#EB7738", label="Smallest", linestyle="--")
ax[0].fill_between(xValuesFirst, lowestCLV, largestCLV, color="#C5DDE3", label="Difference")
#ax[0].set_yscale("log", base=10)
ax[0].set_title("5 Largest and Smallest CLV", fontsize=12, color="black")
ax[0].legend(loc="upper right")

ax[1].scatter(xValuesSecond, custLifetimeValue["Avg. CLV"].values, s=10, c="#DE68DA", ec="black")
ax[1].set_title("Total Avg. CLV", fontsize=12)


plt.show()
```

### Grouping by Customer Category to see: Avg. spend per transaction, Preferred payment methods, Discount sensitivity


```python
# first
custCategoryAvgSpend = df.groupby("Customer Category")["Total Cost"].agg("mean")
custCategoryAvgSpend = custCategoryAvgSpend.round(2)

# second
custCategoryPaymentMethods = df.groupby("Customer Category")["Payment Method"].value_counts()
custCatPaymMethodsDf = custCategoryPaymentMethods.to_frame()
custCatPaymMethodsDf.rename(columns={"count": "Count"}, inplace=True)

# third
custCategoryDiscountsApplied = df.groupby("Customer Category")["Discount Applied"].value_counts()
custDiscAppliedDf = custCategoryDiscountsApplied.to_frame()
custDiscAppliedDf.rename(columns={"count": "Count"}, inplace=True)

# plotting
custCategories = custCategoryDiscountsApplied.index.levels[0]
discountsApplied = {
    "Yes": custDiscAppliedDf.loc[:, True, :].values.flatten(),
    "No": custDiscAppliedDf.loc[:, False, :].values.flatten(),
}

fig, ax = plt.subplots(figsize=(12, 8), constrained_layout=True)

x = np.arange(len(custCategories))
width = 0.25
multiplier = 0

for discount, value in discountsApplied.items():
  offset = width * multiplier
  rects = ax.bar(x + offset, value, width, label=discount, ec="black")
  ax.bar_label(rects, padding=3)
  multiplier += 1

ax.set_ylabel("Count", fontsize=15)
ax.set_xticks(x + width, custCategories)
ax.legend(loc="upper left", ncols=3)
ax.set_title("Discount Sensitivity", fontsize=20)
plt.show()
```

### Seasonal Trends: Total Items per Season

```python
itemsPerSeasonWinter = df.groupby("Season").get_group("Winter")["Total Items"].values
itemsPerSeasonSpring = df.groupby("Season").get_group("Spring")["Total Items"].values
itemsPerSeasonSummer = df.groupby("Season").get_group("Summer")["Total Items"].values
itemsPerSeasonAutumn = df.groupby("Season").get_group("Fall")["Total Items"].values
seasons = ["Winter", "Spring", "Summer", "Autumn"]
items = [itemsPerSeasonWinter, itemsPerSeasonSpring, itemsPerSeasonSummer, itemsPerSeasonAutumn]
colors = ["#fca5a5", "#fde047", "#5eead4"]

plt.rcParams["font.family"] = "monospace"
plt.figure(figsize=(10, 6), dpi=80)

bPlot = plt.boxplot(items, patch_artist=True, tick_labels=seasons, meanline=True)

for patch, color in zip(bPlot["boxes"], colors):
  patch.set_facecolor(color)

plt.title("Items Don't Change Much During The Year", fontsize=14)
plt.tick_params(axis="both", labelsize=12)
```
